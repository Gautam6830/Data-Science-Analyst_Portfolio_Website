Here are some of the available ML model options:
**1. Linear Regression:**
- Linear regression is a simple and interpretable model that assumes a linear relationship between the features and the target variable (salary). It's a good starting point and can provide baseline performance.

**2. Ridge Regression and Lasso Regression:**
- Ridge and Lasso regression are regularization techniques that can be used to handle multicollinearity and prevent overfitting. They are variants of linear regression that add regularization terms to the cost function.

**3. Decision Trees:**
- Decision tree-based models, like Random Forest and Gradient Boosting and AdaBoostRegressor , are capable of capturing non-linear relationships in the data. They can handle both numerical and categorical features and automatically deal with feature importance.

**4. K-Nearest Neighbors (KNN):**
- KNN is a non-parametric method that makes predictions based on the average of the 'k' nearest data points. It can be effective for small to medium-sized datasets.

**5. Polynomial Regression:**
- Polynomial regression can be used to capture non-linear relationships by introducing polynomial features.

```python
# Extract the data from "Tier 1", "Tier2", "Tier 3" college Column List from 'College' DataFrame
# and store all of them in a separate lists 'Tier1', 'Tier2', 'Tier3' named as for further analysis and process

Tier1 = college["Tier 1"].tolist()
Tier2 = college["Tier 2"].tolist()
Tier3 = college["Tier 3"].tolist()
```

```python
# Assign the Tier colleges Weightage in the DataFrame on their Tier based classification
#  If a college is in 'Tier1', set its value to weightage 1
# and If a college is in 'Tier2', set its value to weightage 2
#  and If a college is in 'Tier3', set its value to weightage 3
# Tier1 college has higher weightage then Tier2 and Teir3.
# it means Tier1 weightage of value1 is higher than Tier2 colleges Weightage of value2, and Tier2 colleges weightage value of 2 is higher than Tier3 colleges og weightage value 3

for item in df.College:
    if item in Tier1:
        df["College"].replace(item,1,inplace=True)
    elif item in Tier2:
        df["College"].replace(item,2,inplace=True)
    elif item in Tier3:
        df["College"].replace(item,3,inplace=True)
```

```python
df.head()
```

```python
# Extracting the lists of metro city from the 'cities' 
# and Print the Data Contains in 'metro_city'

metro_city = cities["Metrio City"].tolist()
metro_city
```

```python
# Extracting the lists of non-metro city from the 'cities' 
# and Print the Data Contains in 'non_metro_city'

non_metro_city = cities["non-metro cities"].tolist()
non_metro_city
```

```python
# Converting the categorical data into numerical data
# if city is metro cities assign value as 1 and city is non metro cities assign value as 0

for item in df.City:
    if item in metro_city:
        df["City"].replace(item, 1,inplace = True)
    elif item in non_metro_city:
        df["City"].replace(item, 0,inplace = True)
```

### Observation
1. Outliers present in Previous CTC column
2. Outliers present in CTC column

```python
# Calculate the Interquartile Range (IQR) for "Previous CTC" column
Q1 = df['Previous CTC'].quantile(0.25)
Q3 = df['Previous CTC'].quantile(0.75)

IQR = Q3 - Q1

# Calculate the lower and upper bounds for each feature

upper_limit = Q3 + (1.5 * IQR)
lower_limit = Q1 - (1.5 * IQR)

# Replace outliers with the lower or upper bound
df[(df['Previous CTC'] < lower_limit) | (df['Previous CTC'] > upper_limit)]
```

#### In the above DataFrame, These are outliers present in "Previous CTC"column. As seen these outliers are not extreme, so in my opinion keeping these data may not affect much on our model.

```python
# Calculate the Interquartile Range (IQR) for "CTC" column
Q1 = df['CTC'].quantile(0.25)
Q3 = df['CTC'].quantile(0.75)

IQR = Q3 - Q1

# Calculate the lower and upper bounds for each feature

upper_limit = Q3 + (1.5 * IQR)
lower_limit = Q1 - (1.5 * IQR)

# Replace outliers with the lower or upper bound
df[(df['CTC'] < lower_limit) | (df['CTC'] > upper_limit)]
```

#### As seen above in the Dataframe, these are some outliers in "CTC" column but they are not as extreme that can make any huge difference while making prediction. Therefore in my opinion keeping those outliers into data is more useful than removing.

### Conclusion on detection of Outliers:
- We have seen that There was no such huge amount of extreme outliers present in our dataset that can make any huge difference in our machine learning model. Also from describe function it is clear that there is no extreme outliers.

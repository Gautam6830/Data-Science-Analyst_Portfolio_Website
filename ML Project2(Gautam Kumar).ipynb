## Data importing And Preprocessing

```python
# import pandas library for data manipulation and analysis
# import numpy library for numerical operation and array processing
# import seaborn libray for data visualization 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset to read the CSV file in to a dataframe 
df = pd.read_csv("C:\\Users\\gauta\\Downloads\\sydney_rain prediction.csv",header=0)
```

```python
# Display first few rows of data to understand the data structures
df.head()
```

```python
# check data missing value
df.isna().sum()
```

```python
# Get the Information about data
df.info()
```

```python
# Iterate through  the columns and fill missing values only for numeric columns via mean
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].mean())
```

```python
# Fill the missing value of Categorical column using mode
df.RainToday = df.RainToday.fillna(df.RainToday.mode())
```

```python
# Check missing value is not present in our data
df.isna().sum() # Here we see that RainToday have contain null value is present in our dataset
```

```python
# Create Dummy Variable to convert categorical columm of "RainToday","RainTomorrow" in to numerical column

df = pd.get_dummies(df, columns=["RainToday","RainTomorrow"],drop_first = True)
```

```python
# display first few row to verify the change
df.head()
```

```python
# Check null values are remove
df.isna().sum()
```

```python
# Get the INformation about data after cleaning
df.info() # here we see all column have same no of count
```

```python
# Use describe function to get descriptive statistical info about numerical columns
df.describe()
```

```python
# importing matplotlib to see the Visualizing graph using each column feature 
import matplotlib.pyplot as plt
# using Historgam we see the graph visualization
df.hist(figsize=(15,12),bins = 15)
plt.title("Features Distribution")
plt.show()
```

```python
# Remove Some Unnecessary columns and define target variable (y) and features (x)

x = df.drop(["RainTomorrow_Yes", "Date", "Location"], axis = 1)
y = df["RainTomorrow_Yes"]

# Import the library for data Splitting
from sklearn.model_selection import train_test_split

# Split the Data in to training and test dataset
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
# Import Neccesary libraries 
# Import the library for data modeling and data evaluation
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Initialize the Decision Tree classifier with randomstate and Fit the Model to Training data 
tree = DecisionTreeClassifier(random_state =42)
tree.fit(x_train,y_train)

# Make predictions on test and training dataset
y_test_pred = tree.predict(x_test)
y_train_pred = tree.predict(x_train)

# Calculate accuracy score and confusion matrix on Test data
tree_accuracy = accuracy_score(y_test, y_test_pred)
tree_cm = confusion_matrix(y_test, y_test_pred)


# Import Neccesary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Initialize the Random Forest classifier with randomstate and Fit the Model to Training data 
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)

# Make predictions on test and training dataset
y_test_pred_rf = rf_model.predict(x_test)
y_train_pred_rf = rf_model.predict(x_train)

# Calculate accuracy score and confusion matrix on Test data
rf_accuracy = accuracy_score(y_test, y_test_pred_rf )
rf_cm = confusion_matrix(y_test, y_test_pred_rf)


# Import Neccesary libraries
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Initialize the Gradient Boosting classifier with randomstate and Fit the Model to Training data 
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(x_train, y_train)

# Make predictions on test and training dataset
y_test_pred_gb = gb_model.predict(x_test)
y_train_pred_gb = gb_model.predict(x_train)

# Calculate accuracy score and confusion matrix for test data
gb_accuracy = accuracy_score(y_test, y_test_pred_gb)
gb_cm = confusion_matrix(y_test, y_test_pred_gb)


# Import Neccesary libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Initialize the Ada Boost classifier with randomstate and Fit the Model to Training data 
ada_model = AdaBoostClassifier(random_state=42)
ada_model.fit(x_train, y_train)

# Make predictions on test and training dataset
y_test_pred_ada = ada_model.predict(x_test)
y_train_pred_ada = ada_model.predict(x_train)

# Calculate accuracy score and confusion matrix on Test data
ada_accuracy = accuracy_score(y_test, y_test_pred_ada)
ada_cm = confusion_matrix(y_test, y_test_pred_ada)


# Import the nescerray libraries
import xgboost as xgb 
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Initialize the XG Boost classifier with randomstate and Fit the Model to Training data 
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(x_train, y_train)

# Make predictions on test and training dataset
y_test_pred_xgb = xgb_model.predict(x_test)
y_train_pred_xgb = xgb_model.predict(x_train)

# Calculate accuracy score and confusion matrix on Test data
xgb_accuracy = accuracy_score(y_test, y_test_pred_xgb)
xgb_cm = confusion_matrix(y_test, y_test_pred_xgb)


# Print the accuracy scores for test dataset
print("   Accuracy Score of Testing Dataset   ")
print()
print(f"Decision Tree Accuracy: {tree_accuracy}")
print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"Gradient Boosting Accuracy: {gb_accuracy}")
print(f"Ada Boost Accuracy: {ada_accuracy}")
print(f"XG Boost Accuracy: {xgb_accuracy}")

print()
print()

#Calculate and Print the accuracy scores for training dataset
print("   Accuracy Score of Training Dataset   ")
print()
print(f"Decision Tree Accuracy: {accuracy_score(y_train,y_train_pred)}")
print(f"Random Forest Accuracy: {accuracy_score(y_train,y_train_pred_rf)}")
print(f"Gradient Boosting Accuracy: {accuracy_score(y_train,y_train_pred_gb)}")
print(f"Ada Boost Accuracy: {accuracy_score(y_train,y_train_pred_ada)}")
print(f"XG Boost Accuracy: {accuracy_score(y_train,y_train_pred_xgb)}")


print()
print()

#Print the confusion matrix for test dataset
print("   Confusion Matrix of Testing Dataset   ")
print()
print(f"Decision Tree Confusion Matrix: {tree_cm}")
print(f"Random Forest Confusion Matrix: {rf_cm}")
print(f"Gradient Boosting Confusion Matrix: {gb_cm}")
print(f"Ada Boost Confusion Matrix: {ada_cm}")
print(f"XG Boost Confusion Matrix: {xgb_cm}")
```

## Perform Machine learning Model With HyperParameters Using GridSearchCV function

#### Ensemble methods technique Using through DecisionTreeClassifier 

- 1.Bagging
- 2.Random Forest
- 3.Boosting 

```python
# import future warning library
import warnings
warnings.filterwarnings("ignore", category =FutureWarning)

# Bagging Classifier
# import library for Bagging Classifier for data modeling and evaluation
from sklearn.ensemble import BaggingClassifier

from sklearn.metrics import classification_report

# import the base classifier and grow a full tree in Bagging model thats 
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth = 5, splitter ="best",min_samples_split=3,random_state =42)

# Initialize the bagging classifier and Train the model
bag_clf = BaggingClassifier(base_estimator = clftree, # Using tree classifier as base estimator
                            n_estimators = 1000, # No of trees used in base estimator in ensemble
                            n_jobs= -1, # we use full processing power of computer
                            bootstrap= True, # samples are drwan with replacement with original dataset  
                            random_state = 42 # control random resampling of originaldataset
                           )
```

```python
# Use parameter grid
params_grid={"n_estimators": [10,20,100],
            "max_features": [4,5,6,7,8,9,10],
            "max_samples":[2,5,10]}

# Perform GridSearchCV
from sklearn.model_selection import GridSearchCV
grid_search_bag = GridSearchCV(bag_clf, params_grid, n_jobs= -1, cv=5, scoring = "accuracy")

# Train model usig GridSearchCV
grid_search_bag.fit(x_train,y_train)
```

```python
best_params_bag = grid_search_bag.best_params_
best_bag_est = grid_search_bag.best_estimator_

print("Best Parameter for: ",best_params_bag )
print("Best Estimator :",best_bag_est)
```

```python
# CAlculate and print accuracy score and confusion matrix for GridsearchCV using Bagging Classifier
grid_bagtest_acu = accuracy_score(y_test, best_bag_est.predict(x_test))
grid_bagtrain_acu = accuracy_score(y_train, best_bag_est.predict(x_train))
grid_bag_cm = confusion_matrix(y_test, best_bag_est.predict(x_test))
print("Bagging Classifier Accuracy Score Using GridSearchCV for test data: ",grid_bagtest_acu)
print("Bagging Classifier Accuracy Score Using GridSearchCV: for train data ",grid_bagtrain_acu)
print("Bagging Classifier Confusion Matrix Using GridSearchCV: ",grid_bag_cm)

print()

print("Classification Report:")
print(classification_report(y_test, best_bag_est.predict(x_test)))
```

```python
# Random Forest Classifier
# Import Random Forest Classifer for data modeling and model performance 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score,recall_score

# import GridSearchCV scikit-learn function that performs hyperparameter tuning by training and evaluating a machine learning model using different combinations of hyperparameters
from sklearn.model_selection import GridSearchCV

# Import the Random Forest CLassifier 
rf_clf = RandomForestClassifier(n_estimators = 1000, # No of trees used in base estimator in ensemble
                            n_jobs= -1, # we use full processing power of computer
                            criterion='gini', # measure the quality of split
                            bootstrap= True, # samples are drwan with replacement with original dataset  
                            random_state = 42 # control random resampling of originaldataset
                           )
# Use parameter grid
params_grid_rf={"n_estimators": [10,20,50],
            "max_features": [4,5,6,7,8,9,10],
            "max_depth": [20,30,40,50],
            "min_samples_split":[2,3,10,20]}

# perform GridSearchCV 
grid_search_rf = GridSearchCV(rf_clf, params_grid_rf, n_jobs= -1, cv=5, scoring = "accuracy")

# Train Model using Hyperparameter
grid_search_rf.fit(x_train,y_train)
```

```python
best_params_rf = grid_search_rf.best_params_
best_rf_est = grid_search_rf.best_estimator_

print("Best Parameter for RandomForest Classifier: ",best_params_rf )
print("Best Estimator for RandomForest Classifier: ",best_rf_est)
```

```python
# Calculate and print accuracy score and confusion matrix for GridsearchCV using Bagging Classifier
grid_rftest_acu = accuracy_score(y_test, best_rf_est.predict(x_test))
grid_rftrain_acu = accuracy_score(y_train, best_rf_est.predict(x_train))
grid_rf_cm = confusion_matrix(y_test, best_rf_est.predict(x_test))
print("RandomForest Classifier Accuracy Score Using GridSearchCV for test data: ",grid_rftest_acu)
print("RandomForest Classifier Accuracy Score Using GridSearchCV for train data: ",grid_rftrain_acu)
print("RandomForest Classifier Confusion Matrix Using GridSearchCV : ",grid_rf_cm)

print()

print("Classification Report:")
print(classification_report(y_test, best_rf_est.predict(x_test)))
```
